```python
r1 = np.random.rand(100000)
r2 = np.random.rand(100000)

print(r1,r2)
print(r1.shape, r2.shape)

resultado = 0

t0 = time.time()

for i in range(100000):
  resultado += r1[i] * r2[i]
t1 = time.time()

print(resultado, (t1-t0) * 1000, "ms")

resultado = 0


t0 = time.time()

resultado = np.dot(r1,r2)
t1 = time.time()

print(resultado, (t1-t0) * 1000, "ms")
```
