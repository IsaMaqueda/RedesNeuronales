```python
import cv2 # OpenCv

import numpy as np

import matplotlib.pyplot as plt
```

```python
image = cv2.imread('DL_Data_sets/images/flamingos.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image_grey, cmap='gray')
```

```python
print(image.shape)
print(image_grey.shape)
```

```python
sub_image = image_grey[:350, :350]
plt.imshow(sub_image, cmap='gray')


sub_image2 = image_grey[150:350, 500:750]
plt.imshow(sub_image2, cmap='gray')
```

```python
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

f, (r_ax, g_ax, b_ax) = plt.subplots(1,3,figsize=(20,10))

r_ax.set_title('Red Channel')
r_ax.imshow(r,cmap = 'gray')


g_ax.set_title('Green Channel')
g_ax.imshow(g,cmap = 'gray')

b_ax.set_title('Blue Channel')
b_ax.imshow(b,cmap = 'gray')
```
