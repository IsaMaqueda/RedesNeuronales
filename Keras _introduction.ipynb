```python
import tensorflow as tf

from tensorflow.keras import models, layers, optimizers

import numpy as np

import matplotlib.pyplot as plt
```

```python
print(f"Tensorflow version : {tf.__version__}")
print(f"Keras version : {tf.keras.__version__}")
```

```python
XOR = np.array([[0,0], [0,1], [1,0], [1,1]])
print(f"XOR data: \n{XOR}\nShape: {XOR.shape}")

XOR_labels = np.array([[0, 1, 1, 0]]).T
print(f"XOR labels: \n{XOR_labels}\nShape: {XOR_labels.shape})")
```

```python
network = models.Sequential()


# Hidden Layer
network.add(layers.Dense(3, activation = 'sigmoid', input_shape=(2,), name='hidden_1'))

# output Layer

network.add(layers.Dense(1, activation='sigmoid', name='output'))

# SDG, learning rate, loss

network.compile(optimizer=optimizers.SGD(learning_rate=0.5), loss='binary_crossentropy')

network.summary()
```

```python
history = network.fit(XOR, XOR_labels, epochs=1000)
```

```python
history_dict = history.history

loss_values = history_dict['loss']

fig, ax = plt.subplots(1,1)

ax.set_xlabel('Epochs')
ax.set_ylabel('Error / Loss')

ax.plot(loss_values)
```

```python
input = XOR
prediction = network.predict(input)

print(input)

print(prediction)

input = np.array([[1,1]])
prediction = network.predict(input)

print(input, input.shape)

print(prediction)
```
